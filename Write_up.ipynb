## **Comparison of the two models**

Both models use SHREC11 (600 objects). HKS (heat kernel signature) of each mesh is calculated and stored. Binary classifier NN is trained on the HKS (heat kernel signature) of 2 chosen groups. Multi-class classifier has been built but needs to be refined further as the loss is not doing as well compared to the binary classifier. The contrastive NN model did not have classifier portion so a binary classifier was made.

The contrastive NN classifier is in the notebook titled: Classifier.ipynb and the binary and multi-class classifier are in the notebook titled: HKS_NN.ipynb. 

For binary classifier, groups 0 and 1 were scrambled, and 80% were used for training and the remaining 20% was reserved for testing. For contrastive NN classifier, 10 objects were taken from groups 15 and 17 to be used for testing and the other 30 objects used for training.
